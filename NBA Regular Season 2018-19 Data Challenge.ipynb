# Who is the highest paid player during the seasion?

```python
def to_a_number(values):
    new_vals = []
    for val in values:
        if val == '-':
            new_vals.append(0)
        else:
            new_vals.append(int(val))
    
    return new_vals

nba_1819['salary_int'] = to_a_number(nba_1819['Salary'])
nba_1819.sort_values(by='salary_int', ascending = False)


# Stephen Curry was the highest paid player at 37,457,154
```
